```python
import sqlite3

# Load dataframe into a SQL database
conn = sqlite3.connect('rodent.db')
df.to_sql('inspections', conn, if_exists='replace', index=False)

# Question 1: Which borough has the highest rat activity rate?
query1 = """
    SELECT 
        BOROUGH,
        COUNT(*) as total_inspections,
        SUM(CASE WHEN RESULT = 'Rat Activity' THEN 1 ELSE 0 END) as rat_activity,
        ROUND(100.0 * SUM(CASE WHEN RESULT = 'Rat Activity' THEN 1 ELSE 0 END) / COUNT(*), 2) as rat_rate_pct
    FROM inspections
    WHERE BOROUGH IS NOT NULL
    GROUP BY BOROUGH
    ORDER BY rat_rate_pct DESC
"""

result1 = pd.read_sql_query(query1, conn)
print("Rat Activity Rate by Borough:")
print(result1)
```

```python
# Question 2: How has rat activity trended year over year?
query2 = """
    SELECT 
        YEAR,
        COUNT(*) as total_inspections,
        SUM(CASE WHEN RESULT = 'Rat Activity' THEN 1 ELSE 0 END) as rat_activity,
        ROUND(100.0 * SUM(CASE WHEN RESULT = 'Rat Activity' THEN 1 ELSE 0 END) / COUNT(*), 2) as rat_rate_pct
    FROM inspections
    WHERE YEAR BETWEEN 2021 AND 2024
    GROUP BY YEAR
    ORDER BY YEAR ASC
"""

result2 = pd.read_sql_query(query2, conn)
print("Rat Activity Trend by Year:")
print(result2)
```

```python
# Question 3: Worst zip code hotspots
query3 = """
    SELECT 
        ZIP_CODE,
        BOROUGH,
        COUNT(*) as total_inspections,
        SUM(CASE WHEN RESULT = 'Rat Activity' THEN 1 ELSE 0 END) as rat_activity,
        ROUND(100.0 * SUM(CASE WHEN RESULT = 'Rat Activity' THEN 1 ELSE 0 END) / COUNT(*), 2) as rat_rate_pct
    FROM inspections
    WHERE ZIP_CODE IS NOT NULL
    GROUP BY ZIP_CODE, BOROUGH
    HAVING total_inspections > 500
    ORDER BY rat_rate_pct DESC
    LIMIT 10
"""

result3 = pd.read_sql_query(query3, conn)
print("Top 10 Worst Zip Code Hotspots:")
print(result3)
```

```python
# Question 4: Initial vs Compliance inspection failure rate
query4 = """
    SELECT 
        INSPECTION_TYPE,
        COUNT(*) as total_inspections,
        SUM(CASE WHEN RESULT = 'Rat Activity' THEN 1 ELSE 0 END) as rat_activity,
        ROUND(100.0 * SUM(CASE WHEN RESULT = 'Rat Activity' THEN 1 ELSE 0 END) / COUNT(*), 2) as rat_rate_pct
    FROM inspections
    WHERE INSPECTION_TYPE IN ('Initial', 'Compliance')
    GROUP BY INSPECTION_TYPE
    ORDER BY rat_rate_pct DESC
"""

result4 = pd.read_sql_query(query4, conn)
print("Initial vs Compliance Inspection Failure Rate:")
print(result4)
```

```python
# Question 5: Most improved borough 2021 vs 2024
query5 = """
    SELECT 
        BOROUGH,
        ROUND(100.0 * SUM(CASE WHEN RESULT = 'Rat Activity' AND YEAR = 2021 THEN 1 ELSE 0 END) /
            NULLIF(SUM(CASE WHEN YEAR = 2021 THEN 1 ELSE 0 END), 0), 2) as rate_2021,
        ROUND(100.0 * SUM(CASE WHEN RESULT = 'Rat Activity' AND YEAR = 2024 THEN 1 ELSE 0 END) /
            NULLIF(SUM(CASE WHEN YEAR = 2024 THEN 1 ELSE 0 END), 0), 2) as rate_2024,
        ROUND(
            100.0 * SUM(CASE WHEN RESULT = 'Rat Activity' AND YEAR = 2021 THEN 1 ELSE 0 END) /
            NULLIF(SUM(CASE WHEN YEAR = 2021 THEN 1 ELSE 0 END), 0) -
            100.0 * SUM(CASE WHEN RESULT = 'Rat Activity' AND YEAR = 2024 THEN 1 ELSE 0 END) /
            NULLIF(SUM(CASE WHEN YEAR = 2024 THEN 1 ELSE 0 END), 0)
        , 2) as improvement_pct
    FROM inspections
    WHERE BOROUGH IS NOT NULL
    GROUP BY BOROUGH
    ORDER BY improvement_pct DESC
"""

result5 = pd.read_sql_query(query5, conn)
print("Most Improved Borough (2021 vs 2024):")
print(result5)
```
